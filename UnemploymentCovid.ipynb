```python
#Importing necessary libraries and displaying the head

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px

df=pd.read_csv("covid_desemprego.csv")
```

```python
print(df.describe())
```

```python
#Check for null values in columns
print(df.isnull().sum())
```

```python
#Rename columns
df.columns=["Region","Area","Date","Frequency","EstUnemploymentEmployed","EstEmployed","EstLabourParticipationRate","Longitude","Latitude"]
df
```

```python
print(df.head())
```

```python
#Depiction using a heatmap
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(9,9))
sns.heatmap(df.corr())
plt.show
```

```python
# unemployment trends based on states
df.columns=["Region","Area","Date","Frequency","EstUnemploymentEmployed","EstEmployed","EstLabourParticipationRate","Longitude","Latitude"]
plt.figure(figsize=(15,10))
plt.title("Unemployment in India")
sns.histplot(x="EstEmployed", hue="Region", data=df)
plt.show()
```

```python
# unemployment rates based on area
df.columns=["Region","Area","Date","Frequency","EstUnemploymentEmployed","EstEmployed","EstLabourParticipationRate","Longitude","Latitude"]
plt.figure(figsize=(15,10))
plt.title("Unemployment in India on Area basis")
sns.histplot(x="EstEmployed", hue="Area", data=df)
plt.show()
```
